```python
import pandas as pd
df = pd.read_csv('Student_Responces.csv')
df.head()
df.describe
```

```python
# Data Cleaning
# Drop unwanted features
df.drop(columns = ['In your opinion, what can educational institutions do to help students choose their career paths more effectively? (Open-ended)'])

```

```python
# Data Cleaning

# Drop rows with 'other'
df = df[df.applymap(lambda x: x != 'Other').all(axis=1)]

# Check for missing values
df.isnull().sum()

# Drop missing values
df.dropna()

# There are duplicates, but they are nescessary to the dataset.
```

```python
# Data Cleaning
# Check datatypes - All are correctly 'object'
df.dtypes
```

```python
# Sub-question 1
# Create pivot table
pivot_gender = df.pivot_table(index='What factors affect your career decision-making process?', 
                              columns='Gender', aggfunc='size')


import matplotlib.pyplot as plt

# Visualize
pivot_gender.plot(kind='bar', color=['pink', 'blue']) 
plt.title('Factors Influencing Career Decision-Making Process by Gender')
plt.ylabel('Number of Participants')
plt.xlabel('Factors')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

```

# Sub-question 1
Influence of External and Internal Factors: What are the most prominent factors that influence an individual's career decision-making process? 

It is clear, given the data, Males are more likely to pick a career based on passion for a field. Females are also likely to pick a career based on passion, but almost equally to choosing a career based on financial stability. Parental expectations are least likely to affect a males decision, whereas job market trends are least likely to affect a females decision.

```python
# Sub-question 2
# Create pivot table

pivot_gender_B = df.pivot_table(index='How_did you_choose_your_career_path? ', 
                                columns='Gender', aggfunc='size')

# Visualize
pivot_gender_B.plot(kind='bar', color=['pink','blue'])
plt.title('Methods of Choosing Career Path')
plt.ylabel('Number of Participants')
plt.xlabel('Methods')
plt.xticks(rotation=45, ha='right')
plt.show()


```

# Sub-question 2
How do individuals typically arrive at their career decisions?

The methods used by males to choose a career path are significantly based on personal interest and passion. Females are closely divided amongst three categories, based on financial stability, personal interests and passion, and talked to friends/family. 

```python
# Sub-question 3
# Create pivot table
pivot_gender_C = df.pivot_table(index='Do you think your current career choice aligns with your skills and strengths?', 
                                columns='Gender', aggfunc='size')

# Visualize

fig, ax = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for Males
ax[0].pie(pivot_gender_C['Male'], labels=pivot_gender_C.index, autopct='%1.1f%%', startangle=90, colors=['blue', 'lightblue'], shadow=True)
ax[0].set_title('Male')

# Pie chart for Females
ax[1].pie(pivot_gender_C['Female'], labels=pivot_gender_C.index, autopct='%1.1f%%', startangle=90, colors=['pink', 'hotpink'], shadow=True)
ax[1].set_title('Female')

# Main title
plt.suptitle('Alignment of Career with Skills and Strengths by Gender')

plt.show()

```

# Sub-question 3
How many believe their current career aligns with their skills and strengths?

Overall, the majority of both males and females believe they have chosen careers that align with their skills and strengths. Although the majority of females believe they have chosen careers that align with their skills and strengths, slightly less than half believe their careers do not align with their skills and stengths. 

# Summary

- Males predominantly choose their careers driven by their passions for a particular field. The primary method for males in deciding their career path is rooted in personal interests and passion.
- Females are inclined to choose careers both due to passion and financial stability. The two motivations are almost equal. Females, on the other hand, display a more varied approach. Their career decisions are influenced almost equally by financial prospects, personal passions, and conversations with friends and family

When it comes to external influences, parental expectations have the least impact on males. For females, the prevailing job market trends play a minimal role in shaping their career decisions. A notable majority in both genders believe their chosen careers align well with their innate skills and strengths. However, a subtle observation among females reveals a division: while many are confident about their career choices, a significant portion (just under half) feels their careers might not be the best representation of their strengths.
